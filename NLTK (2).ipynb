```python
#removing all the duplicate data and punctuations
st=set(tokenize)
l=list(st)
s2=" ".join(l)
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
st1 = ""
for char in s2:
    if char not in punctuations:
        st1 = st1 + char
print(st1)
   
```

```python
#again creating tokens
t = word_tokenize(st1)
t
```

```python
#importing porterstemmer and applying stemming to the tokens and creating a new final list
#t.append('am')
#t.append('wow')
#t.append('great')
#t.append('feeling')
#t.append('hate')
#t.append('vita')
#t.append('cdac')
#t.append('sm vita')
from nltk.stem import PorterStemmer
porter = PorterStemmer()
stem=[]
for i in t:
    stem.append(porter.stem(i))
print(stem)
    
```

```python
#doing the whole process again for user's input
n=input("Say Something  : ")
n1=n.lower()
n1t = word_tokenize(n1)
n2=set(n1t)
l1=list(n2)
ns=" ".join(l1)
nts=""
for char in ns:
    if char not in punctuations:
        nts = nts + char
token = word_tokenize(nts)
'''for i in stop_words:
    if i in token:
        token.remove(i)'''
nstem=[]

#comparing two lists to identify whether the user's input is small talk or not
for i in token:
    nstem.append(porter.stem(i))
c=0
for i in nstem:
    if i not in stem:
        break
    else:
        c+=1
if c==len(nstem):
    print("Small Talk")
else:
    print("It is not small talk")
        

```
